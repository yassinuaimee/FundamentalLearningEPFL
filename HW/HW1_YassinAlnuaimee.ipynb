### 1.1

We have :
$$ P_{\lambda}\left(x_{i}\right)= \begin{cases}\frac{e^{-x_{i} / \lambda}}{Z(\lambda)} & \text { if } 1 \leq x_{i} \leq 20 \\ 0 & \text { otherwise }\end{cases}$$


We know that for every probability density function, we have: $\int_{\mathbb{R}} P_{\lambda}(x_i)=1 \Leftrightarrow \int_{ 1 }^{20} P_{\lambda}(x_i) = 1$ $\Leftrightarrow \int_{ 1 }^{20} \frac{e^{-x_{i} / \lambda}}{Z(\lambda)} dx= 1  $ 
$\Leftrightarrow Z(\lambda) = \int_{ 1 }^{20} e^{-x_{i} / \lambda} dx = \lambda (e^{-1/\lambda} -  e^{-20/\lambda})$

$$ \mathbb{E}[X] = \int_{1}^{20} x_i \frac{e^{-x_{i} / \lambda}}{Z(\lambda)} dx = \frac{1}{Z(\lambda)} \int_{1}^{20} x_i e^{-x_{i} / \lambda} dx$$
Moreover 
$$ Z'(\lambda) = \int_{1}^{20} \frac{x_i}{\lambda^2} e^{-x_{i} / \lambda} dx = \frac{1}{\lambda^2} \int_{1}^{20} x_i e^{-x_{i} / \lambda} dx $$ 
Therefore : 
$$ \mathbb{E}[X] = \frac{\lambda^2 \int_{1}^{20} \frac{x_i}{\lambda ^ 2} e^{-x_{i} / \lambda} dx}{Z(\lambda)} = \lambda^2 \frac{Z'(\lambda)}{Z(\lambda)} = \lambda^ 2 \partial_{\lambda} \log Z(\lambda)  = \frac{\lambda(e^{\frac{19}{\lambda}} -1 ) + e^{\frac{19}{\lambda}} - 20 }{e^{\frac{19}{\lambda}} - 1} $$

On another hand $$ \text{var}[X] = \mathbb{E}[X^2] - \mathbb{E}[X]^2 $$

$$  \mathbb{E}[X^2] = \int_{1}^{20} x_{i}^2 \frac{e^{-x_{i} / \lambda} dx} {Z(\lambda)}  $$

$$  \mathbb{E}[X]^2 = \left( \int_{1}^{20} x_{i} \frac{e^{-x_{i} / \lambda} dx} {Z(\lambda)} \right)^2  $$

Therefore we have : 
$$  \text{var}[X] =  \int_{1}^{20} x_{i}^2 \frac{e^{-x_{i} / \lambda} dx} {Z(\lambda)} - \left( \int_{1}^{20} x_{i} \frac{e^{-x_{i} / \lambda} dx} {Z(\lambda)} \right)^2 $$

$$ \Leftrightarrow  \text{var}[X] = \frac{1}{Z(\lambda)} \left( \int_{1}^{20} x_{i}^2 e^{-x_{i} / \lambda} - \frac{\left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} dx \right)^2  } {Z(\lambda)} dx \right)$$

$$ \Leftrightarrow  \text{var}[X] = \frac{\lambda^2}{Z(\lambda)} \left( \int_{1}^{20} \frac{1}{\lambda^2} x_{i}^2 e^{-x_{i} / \lambda} - \frac{1}{\lambda^2} \frac{\left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right)^2 } {Z(\lambda)} \right)$$

$$ \Leftrightarrow  \text{var}[X] = \frac{\lambda^2}{Z(\lambda)} \left( \int_{1}^{20} \frac{1}{\lambda^2} x_{i}^2 e^{-x_{i} / \lambda} - \frac{1}{\lambda^2} \frac{\left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) } {Z(\lambda)} \left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) \right)$$

 and We know from  that $ Z'(\lambda) = \int_{1}^{20} \frac{x_i}{\lambda^2} e^{-x_{i} / \lambda} = \frac{1}{\lambda^2} \int_{1}^{20} x_i e^{-x_{i} / \lambda}$ 

$$ \Leftrightarrow  \text{var}[X] = \frac{\lambda^2}{Z(\lambda)} \left( \int_{1}^{20} \frac{1}{\lambda^2} x_{i}^2 e^{-x_{i} / \lambda} - \frac{ Z'(\lambda) } {Z(\lambda)} \left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) \right)$$

$$ \Leftrightarrow  \text{var}[X] =  \lambda^2  \left( \frac{1}{Z(\lambda)} \int_{1}^{20} \frac{1}{\lambda^2} x_{i}^2 e^{-x_{i} / \lambda} - \frac{ Z'(\lambda) } {Z(\lambda)^2} \left( \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) \right)$$

Knowing that $\partial_{\lambda} \left(\int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) = \int_{1}^{20} \frac{1}{\lambda^2} x_{i}^2 e^{-x_{i} / \lambda} $ and $ \left(\frac{1}{Z(\lambda)}\right)' = - \frac{ Z'(\lambda) } {Z(\lambda)^2} $.

By a variable change $(u =  \int_{1}^{20} x_{i} e^{-x_{i} / \lambda}, v = \frac{1}{Z(\lambda)})$ we can see that ($' $ in according to $\lambda$): 

$$ \Leftrightarrow  \text{var}[X] = \lambda^2 (v'u + u'v)  $$

$$ \Leftrightarrow  \text{var}[X] = \lambda^2 (uv)'  $$

$$ \Leftrightarrow  \text{var}[X] =  \lambda^2 \partial_\lambda \left(\frac{1}{Z(\lambda)} \int_{1}^{20} x_{i} e^{-x_{i} / \lambda} \right) $$

$$ \Leftrightarrow  \text{var}[X] =  \lambda^2 \partial_\lambda \mathbb{E}[X] $$

$$ \Leftrightarrow  \text{var}[X] =  \frac{ \left(e^{\frac{19}{\lambda}} -1 \right)^2  \lambda^2 - 361 e^{\frac{19}{\lambda}}}{ \left(e^{\frac{19}{\lambda}} -1 \right)^2 } $$

```python
def Calc_Mean(lamda):
    return(lamda*(np.exp(19/lamda)-1)+np.exp(19/lamda)-20)/(np.exp(19/lamda)-1)

def Calc_Var(lamda):
    return ((np.exp(19/lamda)-1)**2 * lamda**2 -361*np.exp(19/lamda))/((np.exp(19/lamda)-1)**2)
```

```python
#computing the mean and variamnce as a function of lamda

lamda=np.arange(1, 101)

Mean=Calc_Mean(lamda)
Var=Calc_Var(lamda)

plt.figure(figsize=(16, 4)) 

plt.subplot(1, 2, 1)  # 1 row, 2 columns, this is the first plot
plt.title('Mean as a function of lambda')
plt.plot(lamda, Mean)
plt.xlabel('$\lambda$')
plt.ylabel('E[X]')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, this is the second plot
plt.title('Variance as a function of lambda')
plt.plot(lamda, Var)
plt.xlabel("$\lambda$")
plt.ylabel('Var[X]')

plt.show()
```

$$ P_{\lambda} (\{{x_i}\}^{n}_{i=1}) = \prod^{n}_{i =1} P_{\lambda} (\{{x_i}\}) \qquad \text{(Independence)}$$
$$ = \prod^{n}_{i =1} \frac{1}{Z(\lambda)} e^{-x_i/\lambda} \mathbb{1}_{[1,20]} [x_i]$$ 

$$ =  \left\{\begin{matrix}
\frac{1}{Z(\lambda)^n} e^{-\frac{1}{\lambda} \sum^{n}_{i=1} x_i}   \text{ , if } 1 \leq x_i \leq 20 
\\ 0 ,\text{ otherwise }
\end{matrix}\right.
$$

Log-likelihood function $\mathfrak{L} (\lambda, \{x_i\}^n_{i=1}) = \log(P_{\lambda} (\{{x_i}\}^{n}_{i=1}))/n $

$$ = \left(-n \log(Z(\lambda)) - \frac{1}{\lambda} \sum^{20}_{i=1} x_i dx\right)/n $$
$$ = -\log(Z(\lambda)) - \frac{1}{\lambda n} \sum^{20}_{i=1} x_i $$
$$ = -\log(Z(\lambda)) - \frac{1}{\lambda n} \sum^{20}_{i=1} x_i$$

```python
def Calc_Z(lamda):
  return lamda* (np.exp(-1/lamda) - np.exp(-20/lamda))

def loglh(lamda, TabX):
  
  Z = lamda* (np.exp(-1/lamda) - np.exp(-20/lamda)) # Normalization factor
  
  loglh = 0
  for x in TabX: loglh+= np.log(np.exp(-x/lamda)/Z)
  
  return loglh
```

```python

def Plot_loglh(n, FixedLamda):
  Z=Calc_Z(FixedLamda)
  # Computing P(x_i)
  X = FixedLamda/Z*np.random.exponential(FixedLamda , n)

  # Keeping only values of x that are in [1,20]
  TabX = X[(X >= 1) & (X <= 20)]

  # Computing the log-likelihood array for various lambdas
  TabLamda = np.linspace(.5, 50, 100)
  loglh_lamda = np.array([loglh(lamda, TabX) for lamda in TabLamda])
  
  return [TabLamda, loglh_lamda, np.exp(loglh_lamda - max(loglh_lamda))]
```

```python
FixedLamda=10
n1=10
n2=20
n3=100

r1=Plot_loglh(n1,FixedLamda)
r2=Plot_loglh(n2,FixedLamda)
r3=Plot_loglh(n3,FixedLamda)

#########################################################
# Create the figure with adjusted size and gridspec
fig, axs = plt.subplots(1, 2, figsize=(14, 4))
axs[0].set_title("Log-likelihood")
# Create the first plot (left subplot)
axs[0].plot(r1[0], r1[1], label='n=10') 
axs[0].plot(r2[0], r2[1], label='n=20') 
axs[0].plot(r3[0], r3[1], label='n=100')
axs[0].axvline(FixedLamda, color="r")
axs[0].set_xlabel("$\lambda$")
axs[0].set_ylabel("$L(\lambda)$")
axs[0].legend()

# Create the second plot (right subplot)
axs[1].set_title("Log-likelihood")
axs[1].plot(r1[0], r1[2], label='n=10')
axs[1].plot(r2[0], r2[2], label='n=20') 
axs[1].plot(r3[0], r3[2], label='n=100')
axs[1].axvline(FixedLamda, color="r")
axs[1].set_xlabel("$\lambda$")
axs[1].set_ylabel('$e^{(L-L_{\max})}$')
axs[1].legend()

# Adjust spacing between subplots for better visualization
plt.tight_layout()

# Show the combined plot
plt.show()

```

We can see that the bigger the $n$ the better results we get for our likelihood function. We therefore get a better likelihood function at $n = 100$, this is what expected in theory where as $n \rightarrow \infty $. That our log-likelihood tend to $- log(Z(\lambda))$.

```python
from scipy.optimize import minimize
```

```python
tab_N = np.logspace(1, 3, num=22, base=10)
lmbda_star = 10
lmbda_ML = []

for N in tab_N:
    X = np.random.exponential(lmbda_star , size = int(N))
    # Keeping only values of x that are in [1,20]
    X_tab = X[(X >= 1) & (X <= 20)]

    lmbda_guess = np.random.uniform(0.05, 50)
    optimum = minimize(lambda lmbda0: -loglh(lmbda0, X_tab), lmbda_guess, method='Nelder-Mead')
    lmbda_ML.append(optimum.x)

fig = plt.figure(figsize=(9,6))
plt.title('$\hat{\lambda}_{ML}$', fontsize=16)
plt.xlabel('$N$', fontsize=14)
plt.ylabel('$\lambda$', fontsize=14)
plt.plot(tab_N,lmbda_ML,"*",color='r')
plt.axhline(lmbda_star, label="$\lambda^{\star}$")
# plt.xscale("log")
plt.legend(fontsize=14);
```

We have 

$  \frac{\partial}{\partial \lambda} \log P_{\lambda}(X) = \frac{\partial}{\partial \lambda} \left[ -\frac{x_i}{\lambda} - \log(Z(\lambda) \right]  
 = \frac{x_i}{\lambda^2} - \partial_{\lambda} \log(Z(\lambda)) 
= \frac{x_i}{\lambda^2} -  \frac{\mathbb{E}[X]}{\lambda^2} $

Therefore :

$ \mathbb{E}\left[ \left( \frac{\partial}{\partial \lambda} \log P_{\lambda}(X) \right) ^2 \right] = \mathbb{E}\left[ \left( \frac{x_i}{\lambda^2} -  \frac{\mathbb{E}[X]}{\lambda^2}  \right)^2  \right] = \mathbb{E}\left[ \frac{1}{{\lambda^4} } \left( X-  \mathbb{E}[X] \right)^2  \right] 
 = \frac{1}{{\lambda^4}}  \mathbb{E}\left[  \left( X-  \mathbb{E}[X] \right)^2  \right] = \frac{var[X]}{\lambda^4}   $

```python
def jeffpost(lmbda, x_tab):
  # square root of the fishier score information
  jp = np.log(np.sqrt(Calc_Var(lmbda)/ lmbda**4))
  # normalization factor
  Z = lmbda * (np.exp(-1/lmbda) - np.exp(-20/lmbda))
  for x in x_tab: jp+= np.log(np.exp(-x/lmbda)/Z)
  return jp
```

```python
from numpy.random import choice
from numpy import percentile
```

```python
# function minimizer
def minimizer_lmbda0(func, lmbda0, N):
  # many realizations about 100
  N_rep = 100
  lmbda_est = np.zeros(N_rep)
  for ir in range(N_rep):

    X = np.random.exponential(lmbda0 , int(N))
    X_tab = X[(X >= 1) & (X <= 20)]

    # maximazing the given function
    lmbda_est[ir] = minimize(lambda y:-func(y, X_tab),np.random.uniform(0,30), method='Nelder-Mead').x

    # adding a threshold value for the estimator
    if(lmbda_est[ir] >= 60):
      lmbda_est[ir] = 60 
  return lmbda_est;
```

```python
N_est = 30
tab_lmbda0 = np.linspace(0,50,N_est+1)[1:]
```

```python
# MSE Estimator
def MSE_est(lmbda0_est,lmbda0):
  N_rep = 100
  N_boot=100
  MSE = (lmbda0_est - np.full(N_rep, lmbda0))**2 
  MSE_mean = MSE.mean()
  MSE_boot = choice(MSE,(N_boot,N_rep))
  MSE_boot_mean = MSE_boot.mean(axis=1)
  CI_min = percentile(MSE_boot_mean, [2.5,97.5])[0]
  CI_max = percentile(MSE_boot_mean, [2.5,97.5])[1]

  return MSE_mean, CI_min, CI_max
```

```python
def plot_MSE_J_CR(N):
  N_rep = 100

  lmbda0_J= np.array([ minimizer_lmbda0(jeffpost, lmbda0, N) for lmbda0 in tab_lmbda0])
  lmbda0_ML= np.array([ minimizer_lmbda0(loglh, lmbda0, N) for lmbda0 in tab_lmbda0])

  MSE_ML_N = np.zeros(N_est)
  CI_min_ML_N = np.zeros(N_est)
  CI_max_ML_N = np.zeros(N_est)
  MSE_J_N = np.zeros(N_est)
  CI_min_J_N = np.zeros(N_est)
  CI_max_J_N = np.zeros(N_est)

  for j,lmbda0 in enumerate(tab_lmbda0):
    MSE, CI_min, CI_max = MSE_est(lmbda0_ML[j], lmbda0)
    MSE_ML_N[j] = MSE
    CI_min_ML_N[j] = CI_min
    CI_max_ML_N[j] = CI_max
    MSE, CI_min, CI_max = MSE_est(lmbda0_J[j], lmbda0)
    MSE_J_N[j] = MSE
    CI_min_J_N[j] = CI_min
    CI_max_J_N[j] = CI_max

  fig = plt.figure(figsize=(9, 6), dpi=80)

  plt.title(f'MSE with n = {N}', fontsize=16)
  plt.xlabel('$\lambda_0$', fontsize=14)
  plt.errorbar(tab_lmbda0,MSE_J_N,yerr=[MSE_J_N-CI_min_J_N, CI_max_J_N-MSE_J_N],marker="",color='g', label='Jeffrey')
  plt.errorbar(tab_lmbda0,MSE_ML_N,yerr=[MSE_ML_N-CI_min_ML_N, CI_max_ML_N-MSE_ML_N],marker="",color='r', label='ML')
  plt.plot(tab_lmbda0, (tab_lmbda0**4)/(N**2*Calc_Var(tab_lmbda0)), label='Cramers-Rao')
  plt.legend(fontsize=14)
```

```python
plot_MSE_J_CR(10)
plot_MSE_J_CR(100)
plot_MSE_J_CR(1000)
```

# 2) Probability bounds and a pooling problem

### 2.1

Firstly

$\mathbb{P}(Z_i = k) = p^k(1-p)^{1-k} \text{ with }  k = {0,1}, \quad \forall i \in \mathbb{1}_{[1,m]} \text { (i.i.d) } $ 


Knowing that $\mathbb{P}(a \geq b ) = \mathbb{P}(e^{\lambda a} \geq e^{\lambda b }) $  for any $ \lambda \geq 0 $.

We have : 
$$ \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) = \mathbb{P}\left( e^{\lambda \frac{1}{m} \sum_{i} Z_i} \geq e^{\lambda (p + \epsilon) } \right) \qquad \forall \lambda \geq 0$$

Let $X = e^{\lambda \frac{1}{m} \sum_{i} Z_i} $, $X$ is a positive random variable.
$e^{\lambda (p+\epsilon)} \doteq  a \geq 0 $ thus $$\mathbb{P}(X \geq a) \leq \frac{\mathbb{E}(X)}{a}$$ (with $a \neq 0$) thanks to Markov's inequality .

Thus $$\mathbb{P}\left(\lambda \sum_{i} Z_i \geq \lambda m (p+\epsilon) ) \right) \leq \frac{\mathbb{E}(Y)}{e^{\lambda m (p+\epsilon)}} $$ with $$Y = \lambda \sum_{i} Z_i $$

Therefore
$$  \mathbb{E}[Y] = \mathbb{E}\left(\lambda \sum_{i} Z_i\right) =  \mathbb{E}\left(e^{\lambda \sum_{i} Z_i }\right)  $$
$$ \Leftrightarrow  \mathbb{E}\left(e^{\lambda \sum_{i} Z_i }\right) = \mathbb{E}\left( e^{\lambda Z_1 }  \right) \cdot \mathbb{E}\left( e^{\lambda Z_2 }  \right) \cdots  \mathbb{E}\left( e^{\lambda Z_{m-1} }  \right)  \cdot \mathbb{E}\left( e^{\lambda Z_m }  \right) \text{ (by independence) }$$
$$ \Leftrightarrow \mathbb{E}\left(e^{\lambda \sum_{i} Z_i }\right) = \prod_{i=1}^{m} \mathbb{E}(e^{\lambda Z_i}) =  \mathbb{E}(e^{\lambda Z_1}) ^m $$
$$ \Leftrightarrow  \mathbb{E}\left(e^{\lambda \sum_{i} Z_i }\right) = (p e^{\lambda} + (1-p)e^0)^m $$
$$ \Leftrightarrow \mathbb{E}\left(e^{\lambda \sum_{i} Z_i }\right) = (p e^{\lambda} + (1-p))^m = \mathbb{E}(Y) $$

Finally $$ \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq  \left( \frac{p e^{\lambda} + (1-p)}{e^{\lambda (p+\epsilon)}} \right) ^m$$

### 2.2



We need the minimize $ g_{p, \epsilon}(\lambda) = \left( \frac{p e^{\lambda} + (1-p)}{e^{\lambda(p+\epsilon)}} \right)^m $.

After solving $ \partial_{\lambda} f_{\lambda, p} = 0 $, we get that the $\lambda $  that minimizes the function is $\lambda^{\star} = \log\left(\frac{(p+\epsilon)(1-p)}{p(1-p-\epsilon)} \right) $.
We plug $\lambda^{\star}$ in $f_{\lambda, p} $ we get : 
$$ f_{\lambda, p}(\lambda^{\star}) = f_{\lambda, p}\left(\log\left(\frac{(p+\epsilon)(1-p)}{p(1-p-\epsilon)} \right)\right) = \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon} \right)^{m}  $$ 

Threfore : 

$$ \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq  \left( \frac{p e^{\lambda} + (1-p)}{e^{\lambda (p+\epsilon)}} \right) ^m$$

$$ \Leftrightarrow  \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq  \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon} \right)^{m}  $$

$$ \Leftrightarrow  \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e^{ \log \left( \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon} \right)^{m} \right) }  $$

$$ \Leftrightarrow  \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e^{  m \log \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon}  \right) }  $$

$$ \Leftrightarrow  \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e^{ -  m  \left( - \log \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon}  \right) \right) }  $$


$$ \Leftrightarrow  \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e^{ -  m  f(p,\epsilon) }  $$

$$ \text{with } f(p,\epsilon) =  - \log \left( \left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon}  \right)  $$

### 2.3



$$ f(p, \epsilon = 0) = - \log \left( \left(\frac{p}{p}\right)^p \left(\frac{1-p}{1-p}\right)^{1-p} \right)$$
If $p \neq 0, p \neq 1$  we have : $ \left(\frac{p}{p}\right)^p = 1 $, and  $ \left(\frac{1-p}{1-p}\right)^p = 1 $. 

Then 
$$f(p, \epsilon = 0) = 0  $$



$$ \frac{\partial}{\partial \epsilon} f(p,\epsilon) = \frac{\partial}{\partial \epsilon}\left[-\log \left(\left(\frac{1-p}{1-p-\epsilon}\right)^{1-p-\epsilon}\left(\frac{p}{p+\epsilon}\right)^{p+\epsilon}\right)\right] $$

$$ = \log\left(\dfrac{1-p}{1-p-\epsilon}\right)-\log\left(\dfrac{p}{p+\epsilon}\right) + \frac{p+\epsilon}{p} - \frac{1-p-\epsilon}{1-p} $$

$ \frac{\partial}{\partial \epsilon} f(p,\epsilon) |_{\epsilon = 0} $ $= \log\left(\dfrac{1-p}{1-p}\right)-\log\left(\dfrac{p}{p}\right) = \log(1) - \log(1) + 1 - 1  = 0 $

$$ f''(p, \epsilon) = \frac{1}{p+\epsilon} + \frac{1}{p} + \frac{1}{1-p-\epsilon} + \frac{1}{1-p} = \frac{1}{p(1-p)} + \frac{1}{(p+\epsilon)(1-p-\epsilon)}  $$

From a calculus analysis, we can see that :
$$ \forall p \quad p(1-p) \leq \frac{1}{2} $$
$$  \Leftrightarrow \quad \frac{1}{p(1-p)} \geq 2 \quad \forall p $$

Moreover we have $\epsilon \geq 0 $: 

Therefore with a variable change ($P = p + \epsilon$) we have $(p+\epsilon)(1-p-\epsilon) = P(1-P)$.

From the same analysis as before, we can see that : 

$$ \forall P \quad (p+\epsilon)(1-p-\epsilon) = P(1-P) \leq  \frac{1}{2}$$
$$  \Leftrightarrow \quad \frac{1}{P(1-P)} = \frac{1}{(p+\epsilon)(1-p-\epsilon)} \geq 2 \quad \forall p $$

Thus : 

$$ f''(p, \epsilon) = \frac{1}{p+\epsilon} + \frac{1}{p} + \frac{1}{1-p-\epsilon} + \frac{1}{1-p} $$
$$ f''(p, \epsilon)  = \frac{1}{p(1-p)} + \frac{1}{(p+\epsilon)(1-p-\epsilon)}  \geq 2 + 2 =4 $$ 

$$ \Leftrightarrow f''(p, \epsilon) \geq 4  \quad \forall \epsilon $$
$$ \Leftrightarrow \frac{\partial}{\partial \epsilon}  f(p,\epsilon) \geq 4  \quad \forall \epsilon $$




### 2.4
 Thanks to Taylor's theorem , we have :
$ f(p,\epsilon) = f(p, 0) + \epsilon f'(p, 0) + \epsilon^2 \frac{f''(p,\tilde{\epsilon})}{2} $, for some unknown $\tilde{\epsilon}$. 

We showed that :
$$ f(p, \epsilon = 0) = 0  $$
$$ \frac{\partial}{\partial \epsilon} f(p,0) = 0 $$
$$ \frac{\partial^2}{\partial \epsilon^2} f(p,\epsilon) = f''(p,\epsilon) \geq 4 \quad \forall \epsilon $$ 

We have : 

$$ f(p,\epsilon) = 0 + \epsilon \cdot 0 +  \epsilon^2 \frac{f''(p,\tilde{\epsilon})}{2} = \epsilon^2 \frac{f''(p,\tilde{\epsilon})}{2} $$ 
since $$ f''(p,\epsilon) \geq 4 \quad \forall \epsilon  $$
for an unknown $\tilde{\epsilon}$ we have $f(p,\epsilon) = \epsilon^2 \frac{f''(p,\tilde{\epsilon})}{2} \geq \epsilon^2 \cdot \frac{4}{2} = 2\epsilon^2 $

Meaning that : 
$$ f(p,\epsilon)  \geq  2\epsilon^2 \quad \forall \epsilon$$

We know that : 

$$ \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e ^{-m f(p, \epsilon)} \quad \text{ from 2. }$$

$$ \text{Since } f(p,\epsilon)  \geq  2\epsilon^2 \quad \forall \epsilon$$

We have : 
$$ f(p,\epsilon)  \geq  2\epsilon^2 \quad \forall \epsilon $$
$$\Leftrightarrow  -m f(p,\epsilon)  \leq - 2 m\epsilon^2$$
$$\Leftrightarrow  e^{-m f(p,\epsilon)}  \leq e^{- 2 m\epsilon^2} \text{ (since $exp(x)$ is a strictly increasing function )}  $$


Finally : 

$$ \mathbb{P}\left( \frac{1}{m} \sum_{i} Z_i \geq p + \epsilon \right) \leq e ^{-2 m \epsilon^2} $$

```python
from scipy.stats import bernoulli
```

```python
def RandomArrayVotes(m,p):
  return bernoulli.rvs(p, size=m)
```

```python
mStar=2758
p=[0.2, 0.5, 0.8]

votes=np.array([RandomArrayVotes(mStar,proba) for proba in p])
print(votes)
```

```python
pEst=np.mean(votes, axis=1)
 
print('estimated p for p=0.2=', pEst[0])
print('estimated p for p=0.5=', pEst[1])
print('estimated p for p=0.8=', pEst[2])

pHat=np.mean(votes)
print(pHat)
```

```python
def quantifyP(m, p, tolerance):
  N = 10000
  p_estimate= 0
  count_under_tolerance = 0

  for _ in range(N):
    Tab_Votes = RandomArrayVotes(m, p)
    p_estimate = np.mean(Tab_Votes)

    if(np.abs(p_estimate - p) <= tolerance):
      count_under_tolerance += 1
  
  return count_under_tolerance/N
```

```python
m=2758
p1=0.2
tolerance=1e-2
print('The probability that p hat is correct within one percent  for m=',m ,'and p=',p1,'is',quantifyP(m, p1, tolerance))

p2=0.5
print('The probability that p hat is correct within one percent  for m=',m ,'and p=',p2,'is',quantifyP(m, p2, tolerance))

p3=0.8
print('The probability that p hat is correct within one percent  for m=',m ,'and p=',p3,'is',quantifyP(m, p3, tolerance))
```
